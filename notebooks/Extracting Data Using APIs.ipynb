#### Import and Setup

```python
import os
import requests as r

# URL
url = 'https://api.data.gov/ed/collegescorecard/v1/schools?id=164924&api_key={0}'.format(os.environ['DATA_GOV_API_KEY'])
```

#### Basic Usage

```python
# Fetch data
result = r.get(url)

# Explore response
print("Status Code: {0}".format(result.status_code))
print("Headers: ")
for header in result.headers:
    print("  " + header + ": " + result.headers[header])
```
